```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
#reading data
url = "https://github.com/robitussin/CCADMACL_EXERCISES/blob/ff41fec98bd270f4bbeedb9b8849ae21fdeadfee/Groceries_dataset.csv?raw=true"
dataset = pd.read_csv(url)

dataset.head()
```

```python
client_basket = dataset.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).tolist()
```

### 1. Apply one-hot encoding

```python
from mlxtend.preprocessing import TransactionEncoder

one_hot_transformer = TransactionEncoder()
one_hot_transformer_ary = one_hot_transformer.fit(client_basket).transform(client_basket)
```

```python
one_hot_transformer_ary_df = pd.DataFrame(one_hot_transformer_ary, columns=one_hot_transformer.columns_)
one_hot_transformer_ary_df.head()
```

```python
df = pd.DataFrame(one_hot_transformer_ary, columns=one_hot_transformer.columns_)
df.head()
```

2. Use the Apriori Algorithm to find the most frequent sets

```python
df = apriori(df, min_support=0.0005, use_colnames=True)
df.sort_values(by='support', ascending=False, inplace=True)
```

3. Display the association rules using the lift metric

```python
df_ar = association_rules(df, metric='lift', min_threshold=1)
df_ar.sort_values(by='lift', ascending=False, inplace=True)
df_ar
```
