## Target
The target variable is the median house value for California districts, expressed in hundreds of thousands of dollars ($100,000).

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
housing = fetch_california_housing(as_frame=True)
df = pd.concat([housing.data, housing.target], axis=1)
```

```python
df_housing_featutres = pd.DataFrame(housing.data, columns=housing.feature_names)
df_housing_target = pd.DataFrame(housing.target, columns=['MedHouseVal'])
```

1. Use any filter method to select the best features

```python
from sklearn.feature_selection import f_classif
threshold = 5
high_score_features = []
feature_scores = f_classif(df_housing_featutres, df_housing_target.values.ravel())
```

```python
for score, feature in sorted(zip(feature_scores[0], df_housing_featutres.columns), reverse=True) [:threshold]:
    high_score_features.append(feature)

df_housing_fc = df_housing_featutres[high_score_features]
df_housing_fc.columns
```

2. Use any wrapper method to select the best features

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE

threshold = 5
model_rfe = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=4)
selector = RFE(model_rfe, n_features_to_select=threshold, step=1)

selector = selector.fit(df_housing_featutres, df_housing_target.values.ravel())
selector_ind = selector.get_support()
df_housing_rfe = df_housing_featutres.iloc[:, selector_ind]
df_housing_rfe.columns
```

3. Use any embedded methood to select the best features

```python
from sklearn.feature_selection import SelectFromModel

model_rfe = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=4)
model_rfe.fit(df_housing_featutres, df_housing_target.values.ravel())

sel_sfm = SelectFromModel(model_rfe)
sel_sfm_index = sel_sfm.get_support()
df_housing_sfm = df_housing_featutres.iloc[:, sel_sfm_index]
df_housing_sfm.columns
```
